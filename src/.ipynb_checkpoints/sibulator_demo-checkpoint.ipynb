```python
data = pd.read_csv('vertical_test.csv').dropna(axis=1)
data.columns = data.iloc[0].values
data = data.drop(0).reset_index(drop=True).transpose()
data.columns = data.iloc[0].values
data.drop(data.index[0], inplace=True)
display(data.head())

data = data.iloc[0]
known_sample = {}
for col in ['CaseNumber', 'Sample']:
    if col in data.index:
        data.drop(col, inplace=True)
for loc in data.index:
    allele = data[loc]
    key = loc.split('.')[0]
    known_sample[key] = known_sample.get(key, []) + [allele]
    
known_sample
```
